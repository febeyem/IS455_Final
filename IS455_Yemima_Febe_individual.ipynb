# A scatter plot showing the relationship between the number of reviews and the review scores for listings, using color to differentiate between cities. This visualization will provide an understanding of how guest satisfaction (as reflected in review scores) correlates with the popularity of listings (as measured by the number of reviews), across different cities.

```python
# Drop rows where review scores are missing
data_for_scatter = df.dropna(subset=['review_scores_rating'])

# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data_for_scatter, x='number_of_reviews', y='review_scores_rating', hue='city', alpha=0.6, palette='viridis')
plt.title('Number of Reviews vs. Review Scores Rating by City')
plt.xlabel('Number of Reviews')
plt.ylabel('Review Scores Rating')
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

# Below was supposed to be a chord diagram to Visualize Relationships between Room Type, City, and Guest Capacity but the labels did not correctly show so this is more of a failed visualization. 

```python
chord_data = df.groupby(['city', 'room_type']).size().reset_index(name='counts')

pivot_chord = chord_data.pivot_table(index='city', columns='room_type', values='counts', aggfunc='sum').fillna(0)

# Create the Chord diagram
chord = hv.Chord(pivot_chord).opts(
    opts.Chord(
        cmap='Category20',
        edge_color=dim('source').str(),  # Color edges based on source city
        node_color=dim('index').str(),  # Color nodes based on city names
        labels='index',  # Show city names as labels
        label_text_font_size='10pt',
        label_text_color='black'
    )
)

# Display the diagram
chord
```

```python
df['price'] = np.exp(df['log_price'])

# Drop rows where any of the necessary information is missing
df.dropna(subset=['price', 'review_scores_rating', 'room_type', 'city'], inplace=True)

# Create a FacetGrid scatter plot with regression lines
g = sns.FacetGrid(df, col="city", row="room_type", margin_titles=True, height=3, aspect=1.5)
g.map(sns.regplot, "review_scores_rating", "price", scatter_kws={'alpha':0.3}, line_kws={'color':'red'})

# Add overall titles and labels
g.set_axis_labels('Review Scores Rating', 'Price ($)')
g.set_titles(col_template="{col_name} City", row_template="{row_name}")
g.fig.subplots_adjust(top=0.9)  # adjust the Figure in g
g.fig.suptitle('Relationship between Price and Review Scores by Room Type and City', fontsize=16)

plt.show()
```
