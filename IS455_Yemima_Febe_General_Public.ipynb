```python
from wordcloud import WordCloud
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/Users/febeyemima/Downloads/Airbnb_Data.csv")
df
```

# Above is a glimpse of what the data looks like which is 74111 rows and 29 columns. Below I have put the column names with their data types.

```python
df.dtypes
```

```python
# Preprocesses amenities data
# Converts amenities strings into a single long string for word cloud generation
amenities_series = df['amenities'].str.replace('[{}]', '', regex=True)
amenities_string = ','.join(amenities_series).replace('"', '')

# generate the word cloud for amenities
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(amenities_string)

# Displays the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # No axis for better visualization
plt.title('Word Cloud of Airbnb Amenities')
plt.show()
```

```python
property_counts = df['property_type'].value_counts().head(9)

# Create the horizontal bar graph
plt.figure(figsize=(15, 12))
property_counts.plot(kind='barh', color='skyblue')
plt.title('Top 9 Airbnb Listings by Property Type')
plt.xlabel('Number of Listings')
plt.ylabel('Property Type')
plt.grid(True)
plt.show()
```
